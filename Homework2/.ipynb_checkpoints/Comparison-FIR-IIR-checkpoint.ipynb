## Introduction
The purpose of this homework is to design an FIR Window filter and an IIR filter and compare their results to better understand their use cases   

### Working Steps
1. Define the time domain shape
2. Define the frequency domain shape
3. Generate the filters
4. Evaluate the filters in frequency domain 

### Given Data
| Filter Type | FIR Window | IIR | fs | fc1 | fc2 | ft | rs | rp |
|---|---|---|---|---|---|---|---|---|
| LP | Boxcar | Elliptic | 44100 | 400 | 1000 | 100 | 45 | 0.1 |

## Implementation

```python
import scipy.signal as dsp
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import scipy.fft as sfft
```

### Setting the Data

```python
fc1 = 4000  # Cut-off Frequency (scaled 10:1)
fc2 = 10000 # Cut-off Frequency (scaled 10:1)
fc = 10000 # Cut-off Frequency (scaled 10:1)  => we will use this one

ft = 1000  # Transition Frequency (scaled 10:1)
rp = 0.1    # pass band
rs = 45   # stop band

fs = 44100 # Hz, the Frequency rate
Nyquist = fs/2 # Nyquist Frequency

w = fc/(Nyquist)  
Wn = 0.435 # Best estimated value

M = int(11*fs/(fc1)) # order of the FIR filter

if M % 2 == 0:
    M += 1
```

```python
plt.close('all')
```

### Generating filter kernel 

```python
impulse = np.zeros(501)
impulse[251] = 1
impulseTime = np.arange(0, len(impulse))/fs

# Boxcar
a = 1
bb = dsp.firwin(M, fc, fs = fs, window = ('boxcar'), pass_zero = 'lowpass')
impulseResB = dsp.filtfilt(bb, a, impulse)

plt.figure()
# plt.grid('on')
style.use('dark_background')
plt.plot(impulseTime, impulse, label = 'impulse')
plt.plot(impulseTime, impulseResB/np.max(impulseResB))
plt.xlabel('Time')
plt.ylabel('Filter Gain')
plt.xlim([0.0045, 0.007])
plt.title('Time Response of FIR Boxcar Window')

# Elliptic
be,ae=dsp.ellip(7,rp, rs, Wn, btype='lowpass')
impulseResE = dsp.filtfilt(be, ae, impulse)

plt.figure()
# plt.grid('on')
style.use('dark_background')
plt.plot(impulseTime, impulse, label = 'impulse')
plt.plot(impulseTime, impulseResE/np.max(impulseResE)) # Make it centered
plt.xlabel('Time')
plt.ylabel('Filter Gain')
plt.xlim([0.0045, 0.007])
plt.title('Time Response of IIR Eliptic')
```

## Evaluating Filter in Frequency Domain

```python
#Ideal transfer function
plt.figure()
style.use('dark_background')
plt.plot([0, fc, fc, Nyquist], [1, 1, 0, 0], label = 'Ideal LP filter') #filter gain pts
fb, Hb = dsp.freqz(bb, a = 1, fs = fs)

# plot frequency domain
plt.plot(fb, np.abs(Hb), linewidth = 2, label = 'LP Boxcar Filter')

# plt.grid('on')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.title('Frequency Response of LP FIR Boxcar Window, M = ' + str(M))
plt.legend()

# IIR Filter using Elliptic
wp=(fc+ft)/(fs/2)
ws=(fc-ft)/(fs/2)

N, w=dsp.ellipord(wp=wp, ws=ws, gpass=rp, gstop=rs, fs=fs)
be,ae=dsp.ellip(N,rp, rs, Wn, btype='lowpass')

fe, He=dsp.freqz(b = be, a = ae, fs=fs)

plt.figure()
plt.plot([0, fc, fc, Nyquist], [1,1,0,0], label = 'Ideal LP filter') #filter gain pts
plt.plot(fe,np.abs(He), linewidth = 2, label = 'LP Elliptic Filter')
# plt.grid('on')

plt.xlabel('Frequency')
plt.ylabel('Amplitude(not dB)')
plt.title('Frequency Response of LP IIR Elliptic, N = '+str(N))
```

## Additional Evaluation (zoomed)
To better evaluate the frequency response within the transition area

```python
# Boxcar
plt.figure()

# Ideal transfer function
plt.plot([0, fc, fc, Nyquist], [1,1,0,0], label = 'Ideal filter') #filter gain pts
plt.axis([fc-ft, fc+ft, 0.0, 1.1])
plt.plot([fc-ft/2, fc-ft/2], [0,2], 'w--', label = 'Transition interval')
plt.plot([fc+ft/2, fc+ft/2], [0,2], 'w--')

# Plot frequency domain
plt.plot(fb, np.abs(Hb), label = 'Actual filter')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Frequency Response of LP FIR Boxcar Window (Zoomed), M = ' + str(M))

# Elliptic
plt.figure()

# Ideal transfer function
plt.plot([0, fc, fc, Nyquist], [1,1,0,0], label = 'Ideal filter') #filter gain pts
plt.axis([fc-ft, fc+ft, 0.0, 1.1])
plt.plot([fc-ft/2, fc-ft/2], [0,2], 'w--', label = 'Transition interval')
plt.plot([fc+ft/2, fc+ft/2], [0,2], 'w--')

# Plot frequency domain
plt.plot(fe, np.abs(He), label = 'Actual filter')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Frequency Response of LP IIR Elliptic (Zoomed), N = ' + str(N))
```

### Generateing a random noisy signal

```python
t = np.arange(0, 0.1, 1/fs)
pnts = len(t)
x = 40 * np.sin(2*np.pi*50*t)
noise = 20 * np.random.randn(pnts)
Nsignal = x + noise
plt.figure(figsize = (25,10))
plt.plot(t, Nsignal, label = 'Signal')
style.use('dark_background')
plt.xlabel('Time', fontsize=20)
plt.ylabel('Amplitude', fontsize=20)
plt.title('Time Domain of a Random Noisy Signal', fontsize = 25)
```

### Applying filter in Time domain

```python
Hz = np.linspace(0, Nyquist, int(np.floor(pnts/2)+1))

# Boxcar
# Applying filter in time domain
filtSig = dsp.filtfilt(b = bb, a = 1, x = Nsignal)

# Plotting
plt.figure(figsize = (25,10))
#plt.xlim([1.5, 2.5])
style.use('dark_background')
plt.plot(t, Nsignal, label = 'Original')
plt.plot(t, filtSig, label = 'Filtered')
plt.legend(fontsize = 20)
plt.xlabel('Time', fontsize=15)
plt.ylabel('Amplitude', fontsize=15)
plt.title('Time Domain of the Signal filtered with LP FIR Boxcar Window', fontsize = 25)

# Elliptic
# Applying filter in time domain
filtSigE = dsp.filtfilt(b = be, a = ae, x = Nsignal)
# Plotting
plt.figure(figsize = (25,10))
style.use('dark_background')
plt.plot(t, Nsignal, label = 'Original')
plt.plot(t, filtSigE, label = 'Filtered')
plt.xlabel('Time', fontsize=15)
plt.ylabel('Amplitude', fontsize=15)
plt.title('Time Domain of the Signal filtered with LP IIR Elliptic', fontsize = 25)
```

### Applying filter in Frequency domain

```python
NsignalX = np.abs(sfft.fft(Nsignal)/pnts) ** 2

# Boxcar
filtSigX = np.abs(sfft.fft(filtSig)/pnts) ** 2
plt.figure()

plt.plot(Hz, NsignalX[0:len(Hz)], label = 'Original')
plt.plot(Hz, filtSigX[0:len(Hz)], label = 'Filtered')
plt.plot([fc, fc],[-1, 1],'--', label = 'Cut-off Frequency')
plt.legend()
plt.xlim([fc-ft, fc+ft])
plt.ylim([-0.1,0.6])
plt.xlabel('Frequency')
plt.title('Freq. Domain of the Signal filtered with LP FIR Boxcar Window')

# Elliptic
filtSigEX = np.abs(sfft.fft(filtSigE)/pnts) ** 2

plt.figure()
plt.plot(Hz, NsignalX[0:len(Hz)])
plt.plot(Hz, filtSigEX[0:len(Hz)])
plt.plot([fc, fc],[-1, 1], '--', label = 'Cut-off Frequency')
plt.xlim([fc-ft, fc+ft])
plt.ylim([-0.1,0.6])
plt.xlabel('Frequency')
plt.title('Freq. Domain of the Signal filtered with LP IIR Elliptic')
```

## Conclusion
After designing and analyzing both *FIR* and *IIR* filters, one can conclude that although *IIR* proved to be faster than *FIR* due to having smaller order number, ***FIR* filter is considered the best option** as it provides higher performance comparing to its counterpart since it is non-recursive and inherently stable.
